### Part1.1: Introduction:

Toronto is the capital city of the Canadian province of Ontario. With a recorded population of more than 6 million, it is the most populous city in Canada and the fourth most populous North American city. In early November, Ontario unveiled a new color-coded level system for individual public health units within the province, and Toronto was placed in the Orange-Level. 

By Mid-November, Toronto saw a record number of new infections, and it was announced that the city would be moving to Red-level. The city enacted further, more stringent restrictions on top of the color-code level. For example, Indoor dining will remain closed for a minimum of 28 days; indoor fitness classes are banned, Gyms can reopen with a maximum of 10 people, as well as other.

As of November 20, according to the new response framework, Toronto is placed in the ”Lockdown” category. And effective November 23, the regulations further changed to include Bars and restaurants as take-out only, with no private gatherings outside of one's household. Non-essential retail will be shuttered, with curbside pickup allowed, and sports and recreation activities must cease. 

With the surge in Covid19 cases (and especially that holiday travel season is in play), the rules and regulations are changing daily. 

## Part1.2: Problem Definition:

As Covid19 cases are on the rise, new public displays are being deployed to encourage added caution and to help enforce rules and regulations as well as social distancing guidlines. These displays are crucial to keep information accessible to open small businesses and customers alike during the holiday season. The electronic displays will post up-to-date information, phone hotlines, and web links to increase public awareness and to keep on-top-of the changed regulations. 

This Capstone project aims to provide a study to help Social Workers, non-profit organizations, and the local government with questions about optimizing the locations of outreach channels. And how to best target Ad  campaigns to increase awareness about current COVID19 status and the changing guidelines to ensue the safety of businesses and customers. 

The purpose of this study is to help necessary authorities with their campaign marketing consideration, and primarily with location priorities to place information displays to maximize the message outreach. And one way to provide that is to utilize the Foursquare venue’s popularity and rating as a promoter for the location priority within each neighborhood. The analysis will give insights on the severity of Covid19 cases per each neighborhood and top suggested locations to place the campaign ads and information screen.
